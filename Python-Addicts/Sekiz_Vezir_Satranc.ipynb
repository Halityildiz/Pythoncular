Merhaba yıllar önce karşılaştığım ama hala çözemediğim bir soru sormak istiyorum: Bu soru sekiz vezir problemi diye bilinir. Sekiz veziri satranç tahtasına birbirini yemeden nasıl dizersiniz?

```python
import random
harfler = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
rakamlar = ['1', '2', '3', '4', '5', '6', '7', '8']
surec = [1, 2, 3, 4, 5, 6, 7, 8, 7, 6, 5, 4, 3, 2, 1]

def saga(text):
    iki, b, c, d = [], 0, 0, 7
    while True:
        bir, a = [], 0
        for y in rakamlar[d:]:
            bir.append(harfler[c] + y)
            a += 1
            c += 1
            if a == surec[b]:
                b += 1
                d -= 1
                c = 0
                break
        iki.append(bir)
        if b == 8:break

    while True:
        for i in harfler[1:]:
            e, g, bir = 0, 0, []
            f = harfler.index(i)
            while e < surec[b]:
                bir.append(harfler[f] + rakamlar[g])
                f += 1
                g += 1
                e += 1
            iki.append(bir)
            b += 1
        if b == len(surec):break

    for i in iki:
        if text in i:return i

def sola(text):
    uc, b, c, d = [], 0, 0, 7
    while True:
        dort, a = [], 0
        for y in rakamlar[d:]:
            dort.append(harfler[::-1][c] + y)
            a += 1
            c += 1
            if a == surec[b]:
                b += 1
                d -= 1
                c = 0
                break
        uc.append(dort)
        if b == 8:break

    while True:
        for i in harfler[::-1][1:]:
            e, g, dort = 0, 0, []
            f = harfler.index(i)
            while e < surec[b]:
                dort.append(harfler[f] + rakamlar[g])
                f -= 1
                g += 1
                e += 1
            uc.append(dort)
            b += 1
        if b == len(surec):break
    
    for i in uc:
        if text in i:return i

def yenenler(cift):
    liste = [cift[0] + i for i in rakamlar] + [i + cift[1] for i in harfler if i != cift[0]]
    for i in saga(cift):
        if i not in liste:
            liste.append(i)
    for i in sola(cift):
        if i not in liste:
            liste.append(i)
    return liste

def vezir(n):
    sayi, count, res, result = 0, 0, [], []
    while sayi < 9:
        c1 = random.choice(harfler)
        c2 = random.choice(rakamlar)
        cift = c1 + c2
        count += 1

        if count == 20000:
            res, sayi, count = [], 0, 0

        if sayi == 0:
            res.append(cift)
            uzak_dur = yenenler(cift)
            sayi = 1
        else:
            if cift not in res and cift not in uzak_dur:
                res.append(cift)
                uzak_dur = uzak_dur + yenenler(cift)
                count = 0
            if len(res) == 8:
                if res not in result:
                    result.append(res)
                    res = []
                    uzak_dur = []
                    count = 0
                    sayi = 0
        if len(result) == n:break
    return result

vezir(92) # maksimum 92 farkli cozumu mevcut
```
